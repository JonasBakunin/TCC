```python
def remover_linhas_identicas(df1, df2): #função para remover linhas duplicadas
    # Combinar os dois DataFrames
    df_combinado = pd.merge(df1, df2, how='outer', indicator=True)

    # Filtrar linhas apenas presentes em um DataFrame
    df1_sem_duplicatas = df_combinado[df_combinado['_merge'] == 'left_only'].drop(columns=['_merge'])
    df2_sem_duplicatas = df_combinado[df_combinado['_merge'] == 'right_only'].drop(columns=['_merge'])

    return df1_sem_duplicatas, df2_sem_duplicatas
```

```python
area_plantada_sem_duplicatas, area_colhida_sem_duplicatas = remover_linhas_identicas(area_plantada_sp, area_colhida_sp)
```
